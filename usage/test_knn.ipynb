```python
from models.neighbors.KNeighborsClassifier import KNeighborsClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import datasets
from sklearn import tree
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
data = datasets.load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1234
)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
accuracy_score(y_test, pred)
```

```python
cmap = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])
```

```python
plt.figure()
plt.scatter(X[:,2], X[:,3], c = y, cmap = cmap, edgecolors="k",s = 20)
plt.show()
```

```python
plt.figure(facecolor="white", edgecolor="purple")

plt.xlabel(data.feature_names[2], c = "red")
plt.ylabel(data.feature_names[3], c = "blue")


plt.grid(True, c = "#FF0000", linestyle = "--")
plt.xticks(c = "red")
plt.yticks(c = "blue")

plt.scatter(X_test[:,2], X_test[:,3], c = pred, cmap = cmap, s = 20)
plt.show()
```
