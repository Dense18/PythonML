```python
from utils.metrics import mse, rmse
from LinearRegression import LinearRegression
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import datasets

import matplotlib.pyplot as plt
```

```python
X, y = datasets.make_regression(n_samples=100, n_features = 1, noise = 20, random_state=100)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
# lr = LinearRegression(max_iters=1000, algo="gd",learning_rate=0.01, batch_size=100,tol=0)
lr = LinearRegression(max_iters=1000, algo="sgd",learning_rate=0.001, batch_size=50,tol=0)
lr.fit(X_train, y_train)
```

```python
fig = plt.figure(figsize=(8,6))
# plt.scatter(X_test[:, 0], y_test, color = "b", marker = "o", s = 30)
plt.scatter(X[:, 0], y, color = "b", marker = "o", s = 30)
plt.plot(X, lr.predict(X), c = "red")
plt.show()
```
