```python
from utils.metrics import mse, rmse
from LinearRegression import LinearRegression
from SGDRegression import SGDRegression
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import datasets

import matplotlib.pyplot as plt
import scheduler as sc
```

```python
X, y = datasets.make_regression(n_samples=100, n_features = 1, noise = 20, random_state=100)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
lr = LinearRegression(fit_intercept=True)
lr.fit(X, y)
```

```python
fig = plt.figure(figsize=(8,6))
# plt.scatter(X_test[:, 0], y_test, color = "b", marker = "o", s = 30)
plt.scatter(X[:, 0], y, color = "b", marker = "o", s = 30)
plt.plot(X, lr.predict(X), c = "red")
plt.show()
```

Using Gradient Descent

```python
# gd = GDRegression(max_iters=500, init_learning=0.9, algo="gd", scheduler_func=scf.exp_decay(0.01), tol = 1e-4, random_state=1)
gd = SGDRegression(max_iters=500, init_learning=0.9, algo="gd", scheduler_func=sc.exp_decay(0.01), tol = 1e-4, random_state=1)
gd.fit(X, y)
```
